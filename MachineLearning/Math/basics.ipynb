### Sympy Test

```python
from sympy import plot, symbols, sin

x = symbols('x')
f = sin(x)
plot(f)
```

```python
from sympy.plotting import plot3d

x, y = symbols('x y')
f = 2*x + 3*y
plot3d(f)
```

### Summation $\sum_{i=1}^{5} 2i$

```python
summation = sum(2*i for i in range(1, 6))
print(summation)
```

### Exponent expression using sympy

```python
x = symbols('x')

# Division equivalent to subtraction of powers
expr = x**2 / x**5
print(expr)

# Multiplication equivalent to addition of powers
expr = x**2 * x**3
print(expr)
```

### Logarithm 
A logarithm is a math function that finds a power of a specific number and base
$$
2^x = 8 \\
\log_2 8 =x
$$

```python
from math import log

x = log(8, 2)
print(x)
```

```python
from sympy import log
x = symbols('x')
f = log(x)
plot(f)
```

### Sigmoid Function
$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

```python
from sympy import exp, latex

x = symbols('x')
f = 1 / (1 + exp(-x))
print(latex(f))
plot(f)
```

```python
from math import exp
x = 1
y = 1/ (1 + exp(-x))
print(y)
```
